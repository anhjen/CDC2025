```python
def clean_branch(branch):
    if pd.isna(branch) or branch == "0":
        return "Civilian"

    branch = branch.replace("(Retired)", "").strip()  # remove "(Retired)"

    if "Air Force" in branch:
        return "US Air Force"
    if "Army" in branch:
        return "US Army"
    if "Coast Guard" in branch:
        return "US Coast Guard"
    if "Marine" in branch:
        return "US Marine Corps"
    if "Navy" in branch:
        return "US Navy"

    return "Civilian"
```

```python
nasa_major_categories = pd.read_csv("nasa_major_categories.csv")
```

```python
nasa_major_categories["Military Branch"] = nasa_major_categories["Military Branch"].apply(clean_branch)
```

```python
nasa_major_categories.to_csv("nasa_clean3.csv", index=False)
```
